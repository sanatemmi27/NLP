```python
# def df_corpus(x):
#     corpus = x["Message"].tolist()
#     final_corpus= []
#     for i in range(len(corpus)):
#         text = word_tokenize(corpus[i].lower()) 
#         text = [t for t in text if len(t)>2]
#         text = [wordnet_lemmatizer.lemmatize(y) for y in text if y not in custom]
#         text = [t for t in text if not any(c.isdigit() for c in t)]
#         #sent = " ".join(text)
    
#     final_corpus.append(text)
#     return final_corpus
def my_tokenizer(s):
    s = s.lower()
    tokens = nltk.tokenize.word_tokenize(s)
    tokens = [t for t in tokens if len(t)>2] #remove words lesser than 2 in length
    tokens = [wordnet_lemmatizer.lemmatize(t) for t in tokens] #lemmatize words
    tokens = [t for t in tokens if t not in custom] #remove stopwords and punctuation
    tokens = [t for t in tokens if not any(c.isdigit() for c in t)] # remove digits
    return tokens
    
```

```python
def clean_text(df, column_1, column_2):
    df = df[[column_1, column_2]]
    text = df[column_1].tolist()
    cleaned_text = []
    for x in text:
        cleaned_text.append(my_tokenizer(x))
    joined_text = []
    for x in cleaned_text:
        joined_text.append(' '.join(x))
    df1 = pd.DataFrame({'Message':joined_text})
    df1["sentiment"] = pd.Series(df[column_2].tolist())
    df1["sentiment"] = df1["sentiment"].str.lower()
    return df1
```

```python
#converting inner list to text to find most common words
def split_string(dflist):
    df_2=[]
    for i in range(len(dflist)):
        df_1 = " ".join(dflist[i])
        df_2.append(df_1)
    
    return df_2
```

```python
final_list = split_string(df_list)
final_list
```

```python
words = [x for i in final_list for x in i.split(' ')]
```

```python
words
```

```python
#Function to plot a dispersion graph
def dispersion(word):
    d = list()
    for i in range(0,len(words)):
        if words[i] == word:
            d.append(i)

    fig, ax = plt.subplots()
    ax.vlines(d, 0, 1, edgecolor="black") # <-- ANSWER
    ax.set_xlim([0, len(words)]) # set the lower and upper limits of graph
    ax.set_xlabel('narrative time')
    ax.set_xticks([0],minor=True) # turn off: ax.set_xticks([])
    ax.set_ylabel(word) # turn off by droping this line
    ax.set_yticks([])
    fig.set_figheight(1)
    plt.show()
```
